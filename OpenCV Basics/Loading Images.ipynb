```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Jupiter.jpg")
cv2.imshow("Jupiter", image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
h, w, d = image.shape
print("Height :", h, "\nWidth :", w, "\nDepth :", d)
```

```python
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Jupiter", grayImage)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
grayImage = cv2.imread("Jupiter.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imshow("Jupiter", grayImage)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image) 
"""The difference in color is because the cv2 library uses B,G,R format and the matplotlib uses RGB format so there is mismatch of colors"""
```
